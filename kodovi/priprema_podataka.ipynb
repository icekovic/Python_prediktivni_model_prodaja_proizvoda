```python
import pandas as pd

#ucitavanje podataka
dataset = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss','hours_per_week', 'native_country', 'salary'], index_col=False, encoding="utf-8", delimiter=',\s*', engine='python')
del dataset['fnlwgt']
dataset.head(5)
```

```python
#provjera missing vrijednosti
for col in dataset :
    print(col + ' = ' + (dataset.loc[dataset[col].astype(str) == "?"][col].count()).astype(str))
```

```python
#uklanjanje missing vrijednosti i provjera
dataset = dataset.loc[dataset['workclass'] != '?']
dataset = dataset.loc[dataset['occupation'] != '?']
dataset = dataset.loc[dataset['native_country'] != '?']

for col in dataset :
    print(col + ' = ' + (dataset.loc[dataset[col].astype(str) == "?"][col].count()).astype(str))
```

```python
#kreiranje dummy varijable za ciljnu
dataset['country'] = dataset['native_country'].apply(lambda x : x.strip())
dataset['country_target'] = dataset['country'].apply(lambda x : 1 if x == "England" else 0)
dataset.head()
```

```python
#spremanje u datoteku
dataset.to_csv('pocetni_podaci.csv', index = False)
```
