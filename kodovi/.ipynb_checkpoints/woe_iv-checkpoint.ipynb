```python
import pandas as pd
from matplotlib.pyplot import figure, show, rc
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import pylab as pl

%pylab inline
pylab.rcParams['figure.figsize'] = (25, 8)
```

```python
dataset = pd.read_csv('podaci_diskretizirani.csv', index_col=False, encoding="utf-8", delimiter=',\s*', engine='python')
dataset.head()
```

```python
#woe funkcija
def woe (data) :
    data.columns = ['category', 'target']

    #positive označava ciljnu varijablu (==Ecuador)
    positive = pd.DataFrame(data[(data.target == 1)].groupby('category').count()).reset_index()
    positive.columns = ['category', 'broj_pozitivnih']

    #negative označava sve ono što nije ciljna varijavla (!= Ecuador)
    negative = pd.DataFrame(data[(data.target == 0)].groupby('category').count()).reset_index()
    negative.columns = ['category', 'broj_negativnih']

    #merge podataka - lakše mapiranje i popravljanje podataka (ukoliko ima vrijednosti s nula pošto nije moguće dijeljenje s nulom)
    merge_data = pd.merge(positive, negative, how = 'outer')
    merge_data.fillna(0.0001, inplace = True)
    merge_data.broj_pozitivnih.replace(0, 0.0001, inplace = True)
    merge_data.broj_negativnih.replace(0, 0.0001, inplace = True)

    #suma pozitivnog i negatvinog
    broj_pozitivnih = merge_data['broj_pozitivnih'].sum()
    broj_negativnih = merge_data['broj_negativnih'].sum()

    #izracun postotka pozitivnog i negativnog
    merge_data['postotak_pozitivnih'] = merge_data['broj_pozitivnih'] / broj_pozitivnih
    merge_data['postotak_negativnih'] = merge_data['broj_negativnih'] / broj_negativnih

    #racunanje WoE
    merge_data['woe'] = np.log(merge_data['postotak_pozitivnih'] / merge_data['postotak_negativnih'])
    
    #racunanje IV
    merge_data['iv'] = (merge_data['postotak_pozitivnih'] - merge_data['postotak_negativnih']) * merge_data['woe']

    return merge_data
```

```python
#iv za svaki stupac
for col in dataset :
    data = woe(dataset[[col,"country_target"]])
    iv = data["iv"].sum()
    if (col != "country_target") and (col != "country") and (col != "native_country"):
        print(col + ":", iv, "\n")
```

```python
#izrada grafova
for col in dataset :
    data = woe(dataset[[col,"country_target"]])
    if (col != "country_target") and (col != "country") and (col != "native_country"):
        df_col = data
        width = 0.9
        plt.bar(df_col.index, df_col.woe, width)
        plt.xticks(df_col.index + width/2.,df_col.category)
        print(col)
        plt.show()
```
